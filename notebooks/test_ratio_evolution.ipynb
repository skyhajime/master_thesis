```python
def get_additional_iteration(r, m):
    """get additional iteration

    Args:
        r (float): non zero ratio
        m (int): matrix size
    """
    return 1/2 * (1-r)/r * ((1-r)*m - 1)
```

```python
import numpy as np

ratios = []
tnr = TestNonzeroRatio(**KWARGS)

degs1 = [10*x for x in range(1, 20, 2)]
for deg in degs1:
    iter = 20 if deg%20==0 else 10
    filename = f'test_deg{deg}_iter{iter}'
    header, results = tnr.read_results_from_csv(filename)
    np_results = 1 - np.array(results)/100
    ratios.append(np_results[:,0])

degs2 = [10*x for x in range(20, 32, 2)]
for deg in degs2:
    filename = f'test_deg{deg}_iter10'
    header, results = tnr.read_results_from_csv(filename)
    np_results = 1 - np.array(results)/100
    ratios.append(np_results[:,0])

degs3 = [10*x for x in range(30, 50, 5)]
for deg in degs3:
    filename = f'test_deg{deg}_iter10'
    header, results = tnr.read_results_from_csv(filename)
    np_results = 1 - np.array(results)/100
    ratios.append(np_results[:,0])
```

```python
plt.figure(figsize=(16,9))
plt.boxplot(ratios, vert=True, patch_artist=True, labels=degs1+degs2+degs3)
plt.ylabel("reduction ratio[%]")
plt.xlabel("matrix column size")
plt.show()
```
